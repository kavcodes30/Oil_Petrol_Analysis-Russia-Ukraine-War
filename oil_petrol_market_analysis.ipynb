```python
crude_oil_df["Volatility"] = (
    crude_oil_df["Brent_Change_Pct"]
    .rolling(7)
    .std()
)
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    crude_oil_df["Phase"],
    crude_oil_df["Volatility"]
)

plt.title("Oil Market Volatility Over Time")
plt.xlabel("Date")
plt.ylabel("Volatility")

plt.show()
```

```python
country_impact_df["Risk_Score"] = (
    country_impact_df["Inflation_Risk"]
    .map({
        "Low":1,
        "Moderate":2,
        "High":3,
        "Very High":4,
        "Extreme":5
    })
)
country_impact_df
```

```python
risk_table = country_impact_df.pivot_table(
    values="Risk_Score",
    index="Country"
)
```

```python
plt.figure(figsize=(8,10))

sns.heatmap(
    risk_table,
    annot=True,
    cmap="Reds"
)

plt.title("Inflation Risk Heatmap")

plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(country_impact_df['Country'],country_impact_df['Oil_Import_Pct'])
plt.plot(country_impact_df['Country'],country_impact_df['Risk_Score'])
```

```python
petrol_prices_df['Percentage_Change']=((petrol_prices_df['Mar7_USD']-petrol_prices_df['Before_War_USD'])/petrol_prices_df['Before_War_USD'])*100
```

```python
plt.figure(figsize=(20,10))
plt.plot(wartimeline_df['Event'].head(14),petrol_prices_df['Percentage_Change'])
plt.xlabel('Events during Conflict')
plt.ylabel('% Change in Price of Petrol (USD)')
plt.title('Events vs % Change in Petrol Prices')
```

```python
avg_brent = crude_oil_df["Brent_USD"].mean()
```

```python
comparison_df = petrol_prices_df.copy()

comparison_df["Avg_Crude_Price"] = avg_brent
```

```python
comparison_df['avg_petrol']=(petrol_prices_df["Before_War_USD"]+petrol_prices_df['Mar7_USD'])/2
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=comparison_df,
    x="Avg_Crude_Price",
    y="avg_petrol",
    hue="Country",
    s=120
)

plt.title("Average Crude Oil Price vs Petrol Prices")
plt.xlabel("Average Crude Oil Price")
plt.ylabel("Petrol Price")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=country_impact_df,
    x="Oil_Import_Pct",
    y="Risk_Score",
    hue="Country",
    s=120
)

plt.title("Oil Dependency vs Inflation Risk")
plt.xlabel("Oil Import Percentage")
plt.ylabel("Risk Score")

plt.show()
```

Conclusion:

This project analyzed the relationship between crude oil prices, petrol prices, import dependency, and economic risk across different countries. Using Python libraries such as Pandas, Matplotlib, and Seaborn, multiple visualizations and statistical analyses were performed to understand global oil market patterns.

Countries with higher oil import percentages generally showed increased vulnerability to market instability and geopolitical disruptions.

Through heatmaps, scatter plots, histograms, and comparative visualizations, the project highlighted:
